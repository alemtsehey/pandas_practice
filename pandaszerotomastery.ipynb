**When should you use .loc[] or .iloc[]?**

Use .loc[] when you're selecting rows and columns based on their lables or a condition (e.g. retrieving data for specific columns).
Use .iloc[] when you're selecting rows and columns based on their integer index positions (e.g. extracting the first ten rows regardless of the labels).

**Group by Make and Calculate the Average Price:
You might want to see the average price of cars by make.**

```python
# Group by 'Make' and calculate the mean 'Price'
grouped = cars.groupby('Make')
prices_by_make = grouped['Price']
print(prices_by_make.mean())
#print(cars.groupby('Make')['Price'].mean())
```

```python
print(cars.groupby('Make')['Price'].mean())
```

1. cars.groupby('Make')
Grouping the DataFrame by 'Make':
The groupby('Make') method is used to group the rows of the DataFrame based on the values in the 'Make' column.
Each unique value in the 'Make' column (e.g., Toyota, Honda, BMW) forms a group.
The result is a GroupBy object, which is a special Pandas object that contains information about the groups but has not yet performed any operations.

2. ['Price']
Selecting the 'Price' Column:
After grouping, ['Price'] is used to select the 'Price' column from each group.
This operation extracts the 'Price' column for each group (e.g., all prices for Toyota cars, all prices for Honda cars, etc.).

3. .mean()
Calculating the Mean:
The .mean() function is then applied to the 'Price' column within each group.
It calculates the average price for each group (each make in this case).
The result is a new Pandas Series where the index is the 'Make' (e.g., Toyota, Honda) and the values are the average prices.

**Aggregate functioins**

Aggregate function in Pandas performs summary computations on data, often on grouped data. But it can also be used on Series objects.

This can be really useful for tasks such as calculating mean, sum, count, and other statistics for different groups within our data.

**single aggregate functions**

```python
import pandas as pd
data = {
    'Category': ['A', 'A', 'B', 'B', 'A', 'B'],
    'Value': [10, 15, 20, 25, 30, 35]
}

df = pd.DataFrame(data)
```

```python
df
```

```python
#calculate total sum, mean, max, min of the value columnn
total_sum = df['Value'].max()
total_sum
```

```python
#or
total = df['Value'].aggregate('max')
total
```

**Multiple aggregate functions**

```python
df
```

```python
result = df.groupby('Category')['Value'].agg(['sum', 'mean', 'min', 'max'])
result
```

```python
result = df.groupby('Category')['Value'].aggregate(['mean']) # mean() without agg functio
result
```

```python
result = df.groupby('Category')['Value'].agg(['sum', 'mean', 'max', 'min'])
result
```
