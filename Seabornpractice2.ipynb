```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data (you can replace this with your own data)
data = {
    'Genre': ['Comedy', 'Drama', 'Action', 'Romance', 'Thriller'],
    'Number of Movies': [25, 18, 15, 12, 10]
}

# Create a DataFrame from the data
df = pd.DataFrame(data)

# Create the barplot
sns.barplot(x='Genre', y='Number of Movies', data=df, palette='viridis', hue = 'Genre')

# Customize the plot
plt.title('Number of Movies by Genre')
plt.xlabel('Genre')
plt.ylabel('Number of Movies')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability

# Display the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample data (replace with your own)
data = {
    'Genre': ['Comedy', 'Drama', 'Action', 'Comedy', 'Drama', 'Action', 'Comedy', 'Drama', 'Action'],
    'Rating': ['G', 'G', 'G', 'PG', 'PG', 'PG', 'R', 'R', 'R'],
    'Number of Movies': [10, 5, 8, 15, 12, 10, 20, 18, 15]
}

df = pd.DataFrame(data)

# Create the barplot with hue
sns.barplot(x='Genre', y='Number of Movies', hue='Rating', data=df, palette='viridis')

plt.title('Number of Movies by Genre and Rating')
plt.xlabel('Genre')
plt.ylabel('Number of Movies')
plt.show()
```
